<h3>터미널의 위도,경도를 바탕으로 터미널 간 거리를 가진 dictionary 생성</h3>
참고: https://stackoverflow.com/questions/19412462/getting-distance-between-two-points-based-on-latitude-longitude?utm_medium=organic&utm_source=google_rich_qa&utm_campaign=google_rich_qa

```python
def distance(dept, arr):
    R = 6373.0 # approximate radius of earth in km 
    
    dlat = abs(dept[0]-arr[0])
    dlng = abs(dept[1]-arr[1])
    
    a = sin(dlat / 2)**2 + cos(dept[0]) * cos(arr[0]) * (sin(dlng / 2))**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R*c
    
    return distance
```

```python
# (숫자 작은 터미널, 숫자 큰 터미널) 만 key 값으로 저장되어있는 distance_dic. 그러나 이 경우 누가 먼저오는지 main dataframe에서 매번 연산해야하는 문제가 있었다. 
# 그래서 (숫자 큰 터미널, 숫자 작은 터미널) 도 key 값으로 만들기로 하였다. (dictionary의 길이가 2배가 된다.)

distance_dic = {}

for idx1, row1 in df.iterrows():
    dept = (row1['lat'], row1['lng'])
    
    for idx2, row2 in df[idx1:].iterrows():
        arr = (row2['lat'], row2['lng'])
        
        dist = round(distance(dept,arr),2)
        distance_dic[(row1['ter_no'],row2['ter_no'])] = dist
        distance_dic[(row2['ter_no'],row1['ter_no'])] = dist
        
```
